```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

# Example: Suppress all warnings
warnings.filterwarnings("ignore")
```

```python
class PhoneDBProcessor:
    def __init__(self):
        self.data = None

    def load_data(self, file_path):
        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                csv_reader = csv.DictReader(file)
                self.data = list(csv_reader)
            print("Data loaded successfully.")
        except FileNotFoundError:
            print(f"File not found at {file_path}")

    def retrieve_device_info(self, oem_id):
        if self.data is not None:
            device_info = []
            for record in self.data:
                if record['oem_id'] == oem_id:
                    device_info.append({
                        'Model Name': record['model'],
                        'Manufacturer': record['manufacturer'],
                        'Weight': record['weight_gram'],
                        'Price': record['price'],
                        'Price Unit': record['price_currency']
                    })
            return device_info
        else:
            print("Data not loaded. Use load_data() method to load data first.")
```

##### b1. Identify the top 5 regions where a specific brand of devices was sold.

```python
class PhoneDBAnalyzer:
    def __init__(self):
        self.data = None

    def load_data(self, file_path):
        try:
            self.data = pd.read_csv(file_path)
            print("Data loaded successfully using pandas.")
        except FileNotFoundError:
            print(f"File not found at {file_path}")

    def top_regions_by_brand(self, brand_name):
        if self.data is not None:
            brand_data = self.data[self.data['brand'] == brand_name]
            top_regions = brand_data['market_regions'].value_counts().head(5)
            return top_regions
        else:
            print("Data not loaded. Use load_data() method to load data first.")

    def plot_brand_distribution(self):
        if self.data is not None:
            brand_distribution = self.data['brand'].value_counts()
            brand_distribution.plot(kind='bar', figsize=(10, 6), color='skyblue')
            plt.title('Distribution of Device Brands')
            plt.xlabel('Brand')
            plt.ylabel('Count')
            plt.show()
        else:
            print("Data not loaded. Use load_data() method to load data first.")
```

```python
# Example usage
if __name__ == "__main__":
    phone_db_analyzer = PhoneDBAnalyzer()

    # Use the file path received from task a)
    file_path_from_task_a = 'C:/Users/device_features.csv'

    # Load data from the CSV file using pandas
    phone_db_analyzer.load_data(file_path_from_task_a)

    # Example: Identify the top 5 regions where a specific brand of devices was sold
    brand_to_analyze = 'Asus'
    top_regions = phone_db_analyzer.top_regions_by_brand(brand_to_analyze)

    # Display the result
    if top_regions is not None and not top_regions.empty:
        print(f"\nTop 5 Regions for {brand_to_analyze} Devices:")
        print(top_regions)
    else:
        print(f"No data available for {brand_to_analyze} devices.")

    # Plot the distribution of device brands
    phone_db_analyzer.plot_brand_distribution()
```

##### c3. Create separate charts illustrating the monthly average price trends (in GBP) for devices released in each year from 2020 to 2023. Each chart should focus on a specific year.

```python
class PhoneDBVisualizer:
    def __init__(self):
        self.data = None

    def load_data(self, file_path):
        try:
            self.data = pd.read_csv(file_path)
            print("Data loaded successfully using pandas.")
        except FileNotFoundError:
            print(f"File not found at {file_path}")

    def monthly_average_price_trends(self, year):
        if self.data is not None:
            # Filter data for the specific year
            year_data = self.data[pd.to_datetime(self.data['released_date']).dt.year == year]
            
            # Convert the 'price' column to strings
            year_data['price'] = year_data['price'].astype(str)

            # Convert the 'price' column to numeric (assuming it contains strings with currency symbols)
            year_data['price'] = pd.to_numeric(year_data['price'].str.replace('', '').str.replace(',', ''), errors='coerce')
            
            # Extract month and year from the release date
            year_data['release_month'] = pd.to_datetime(year_data['released_date']).dt.month

            # Group by month and calculate the average price
            monthly_avg_prices = year_data.groupby('release_month')['price'].mean()

            # Plot the chart with different line styles and colors
            line_styles = ['-', '--', '-.', ':']
            line_colors = ['blue', 'green', 'red', 'purple']

            plt.figure(figsize=(10, 6))
            plt.plot(monthly_avg_prices.index, monthly_avg_prices.values, marker='o', linestyle=line_styles[year % len(line_styles)], color=line_colors[year % len(line_colors)])
            plt.title(f'Monthly Average Price Trends in GBP for Devices Released in {year}')
            plt.xlabel('Month')
            plt.ylabel('Average Price (GBP)')
            plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
            plt.grid(True)
            plt.show()
        else:
            print("Data not loaded. Use load_data() method to load data first.")
```

```python
# Example usage
if __name__ == "__main__":
    phone_db_visualizer = PhoneDBVisualizer()

    # Use the file path received from task a)
    file_path_from_task_a = 'C:/Users/device_features.csv'

    # Load data from the CSV file using pandas
    phone_db_visualizer.load_data(file_path_from_task_a)

    # Example: Create separate charts for each year from 2020 to 2023
    for year in range(2020, 2024):
        phone_db_visualizer.monthly_average_price_trends(year)
```
